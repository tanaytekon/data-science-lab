### Question 3)

Find and remove outliers by analyzing the above output of question 2. Explain your outlier detection process.

```python
# solution code
import numpy as np

#################### 'y' attribute ########################################
fig, ax = plt.subplots(1,2,figsize=(25,4))

# Trying IQR method
Q1 = np.percentile(df_jewel['y'], 25, method='midpoint')
Q3 = np.percentile(df_jewel['y'], 75, method='midpoint')
IQR = Q3 - Q1
df_jewel_iqr = df_jewel.loc[(df_jewel['y']<(Q3+1.5*IQR)) & (df_jewel['y']>(Q1-1.5*IQR))]
ax[0].plot(df_jewel_iqr['price'], df_jewel_iqr['y'])
ax[0].set_xlabel('price')
ax[0].set_ylabel('y')
ax[0].set_title("'y' values after outlier removal by IQR method")
# As seen in the graph, IQR method removed some values on upper part which are not outliers.
# These values are higher than expected, but since they are not too extreme, they cannot be considered as outliers.

# Since IQR method did not provide reliable results and outliers can seem from graph,
# it is not necessary to use other methods like z-score and standart deviation, so manual removal will be more reliable.

# 'y' values after manual removal.
df_jewel_manual = df_jewel.loc[(df_jewel['y']<(10))&(df_jewel['y']>(2))]
ax[1].plot(df_jewel_manual['price'], df_jewel_manual['y'])
ax[1].set_xlabel('price')
ax[1].set_ylabel('y')
ax[1].set_title("'y' values after manual removal")
plt.show()
# As seen in the graph, manual removal worked successfully for 'y' attribute
df_jewel = df_jewel_manual


#################### 'z' attribute ########################################
fig, ax = plt.subplots(1,2,figsize=(25,4))

# Trying IQR method
Q1 = np.percentile(df_jewel['z'], 25, method='midpoint')
Q3 = np.percentile(df_jewel['z'], 75, method='midpoint')
IQR = Q3 - Q1
df_jewel_iqr = df_jewel.loc[(df_jewel['z']<(Q3+1.5*IQR))&(df_jewel['z']>(Q1-1.5*IQR))]
ax[0].plot(df_jewel_iqr['price'], df_jewel_iqr['z'])
ax[0].set_xlabel('price')
ax[0].set_ylabel('z')
ax[0].set_title("'z' values after outlier removal by IQR method")
# As seen in the graph, IQR method removed some values on upper part which are not outliers.
# These values are higher than expected, but since they are not too extreme, they cannot be considered as outliers.
# Also, IQR method could not remove two obvious outliers which are below 2(That is not so important because they can be removed by hand).

# Similarly, manual removal will be more reliable.

# 'z' values after manual removal.
df_jewel_manual = df_jewel.loc[(df_jewel['z']<(7))&(df_jewel['z']>(2))]
ax[1].plot(df_jewel_manual['price'], df_jewel_manual['z'])
ax[1].set_xlabel('price')
ax[1].set_ylabel('z')
ax[1].set_title("'z' values after manual removal")
plt.show()
# As seen in the graph, manual removal worked successfully for 'z' attribute
df_jewel = df_jewel_manual


#################### 'depth' attribute ########################################
fig, ax = plt.subplots(1,2,figsize=(25,4))

# Trying IQR method
Q1 = np.percentile(df_jewel['depth'], 25, method='midpoint')
Q3 = np.percentile(df_jewel['depth'], 75, method='midpoint')
IQR = Q3 - Q1
df_jewel_iqr = df_jewel.loc[(df_jewel['depth']<(Q3+1.5*IQR))&(df_jewel['depth']>(Q1-1.5*IQR))]
ax[0].plot(df_jewel_iqr['price'], df_jewel_iqr['depth'])
ax[0].set_xlabel('price')
ax[0].set_ylabel('depth')
ax[0].set_title("'depth' values after outlier removal by IQR method")
# As seen in the graph, IQR method removed some values on upper and lower part which are not outliers.
# These values are higher or lower than expected, but since they are not too extreme, they cannot be considered as outliers.

# Similarly, manual removal will be more reliable.

# 'depth' values after after manual removal.
df_jewel_manual = df_jewel.loc[(df_jewel['depth']<(75)) & (df_jewel['depth']>(50.8))]
ax[1].plot(df_jewel_manual['price'], df_jewel_manual['depth'])
ax[1].set_xlabel('price')
ax[1].set_ylabel('depth')
ax[1].set_title("'depth' values after manual removal")
plt.show()
# As seen in the graph, manual removal worked successfully for 'depth' attribute.
df_jewel = df_jewel_manual


#################### 'table' attribute ########################################
fig, ax = plt.subplots(1,2,figsize=(25,4))
# Trying IQR method
Q1 = np.percentile(df_jewel['table'], 25, method='midpoint')
Q3 = np.percentile(df_jewel['table'], 75, method='midpoint')
IQR = Q3 - Q1
df_jewel_iqr = df_jewel.loc[(df_jewel['table']<(Q3+1.5*IQR))&(df_jewel['table']>(Q1-1.5*IQR))]
ax[0].plot(df_jewel_iqr['price'], df_jewel_iqr['table'])
ax[0].set_xlabel('price')
ax[0].set_ylabel('table')
ax[0].set_title("'table' values after outlier removal by IQR method")
# As seen in the graph, IQR method removed some values on upper and lower part which are not outliers.
# These values are higher or lower than expected, but since they are not too extreme, they cannot be considered as outliers.

# Similarly, manual removal will be more reliable.

# 'table' values after after manual removal.
df_jewel_manual = df_jewel.loc[(df_jewel['table']<(75)) & (df_jewel['table']>(45))]
ax[1].plot(df_jewel_manual['price'], df_jewel_manual['table'])
ax[1].set_xlabel('price')
ax[1].set_ylabel('table')
ax[1].set_title("'table' values after outlier removal by hand")
plt.show()
# As seen in the graph, manual removal worked successfully for 'table' attribute.
df_jewel = df_jewel_manual
```

### Question 4)
Replace ordinal values (cut, color, clarity) according to rules below:
- For "cut": 'Good':1, 'Very Good':2, 'Fair':3, 'Ideal':4, 'Premium':5
- For "color": 'J':1, 'I':2, 'H':3, 'G':4, 'F':5, 'E':6, 'D':7
- For "clarity": 'I1':1, 'SI1':2, 'SI2':3, 'VS1':4, 'VS2':5, 'VVS1':6, 'VVS2':7, 'IF':8

### Question 6)
Shuffle the data points and split the dataset into training and testing set, with a 0.7-0.3 ratio (training set will be %70, and test set will be %30 of the dataset) (Write your own shuffling and spliting code here, do not use predefined function train_test_split).



```python
# solution code
# shuffling
df_jewel = df_jewel.iloc[np.random.permutation(len(df_jewel))]
df_jewel.reset_index(drop=True, inplace=True)

# splitting
trainMask = [True]*round(len(df_jewel)*7/10)+[False]*(len(df_jewel)-round(len(df_jewel)*7/10))
train = df_jewel[trainMask]
test = df_jewel[~np.array(trainMask)]
```

### Question 7)

Create an instance of kNN with a number of neighbors between 1-30, then fit kNN regression model to train data. Then, make prediction and calculate mean squared error (MSE). Calculate MSE for k value between 1-30 and print them. Finally, create elbow function to find best k value and plot the elbow curve. Explain how you find the best k-value.

```python
# solution code
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

X_train = train.drop('price',axis=1).values
X_test = test.drop('price',axis=1).values

y_train = train['price'].values
y_test = test['price'].values

mse_values= []
for k in range(1,31):
    model = KNeighborsRegressor(n_neighbors=k)
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    mse_values.append(mean_squared_error(y_test,prediction))
    
fig, ax = plt.subplots()
ax.plot(list(range(1,31)), mse_values)
ax.set_ylabel('Mean Squared Error')
ax.set_xlabel('k value')
ax.ticklabel_format(style='plain')
plt.show()

# I put mean square values for every k to the mse_values list.
# Then, I took minimum MSE value from that list and I found the matching k value for this minumum MSE error by using list's index.
min_mse = min(mse_values)
best_k = mse_values.index(min(mse_values))+1
print("Minimum MSE:",min_mse)
print("Best k value:",best_k)
```

### Question 8)

Run 10-Fold Cross validation (CV) and print accuracy values for each run of CV. Also print the final average accuracy of the model.

```python
# solution code
from sklearn.model_selection import cross_val_score

model = KNeighborsRegressor(n_neighbors=best_k)
X = df_jewel.drop('price', axis=1).values
y = df_jewel['price'].values
cv_scores = cross_val_score(model, X, y, cv=10)

print("Accuracy values for each run:")
for i in range(10):
    print(f"{i+1}:", cv_scores[i])

print()
print(f"Final average accuracy: {np.mean(cv_scores)}")
```
