```python
import numpy as np
import pandas as pd
from sqlite3 import connect
import sqlite3
import warnings
warnings.filterwarnings('ignore')
```

### Question 1)
Import every 10th row from given dataset as a dataframe. Print obtained dataframe.

```python
# solution code
df = pd.read_csv("BostonHousing.csv").iloc[::10,:]
display(df)
```

### Question 2)
Use "age" attribute to crate array and print mean, minimum, maximum value of age


```python
# solution code
ageArray = df['age'].to_numpy()
mean = np.mean(ageArray)
minimum = np.min(ageArray)
maximum = np.max(ageArray)
print(f"Mean: {mean}\nMinimum: {minimum}\nMaximum: {maximum}")
```

### Question 3) 

Show "value of rm" > "average of rm value" and "age" for 20 records.

Note: RM- average number of rooms per dwelling


```python
# solution code
averageRm = np.mean(df['rm'].to_numpy())
df['value of rm > average of rm value'] = np.where(df['rm'] > averageRm, True, False)
display(df[['value of rm > average of rm value','age']].head(20))
df.drop('value of rm > average of rm value', inplace=True, axis=1)
```

### Question 4)
Print the row of the third largest age value from column 'age' (keep first unique value) from the dataframe created in the first question.


```python
# solution code
age_df = df.drop_duplicates(subset='age', keep='first')
age_df.sort_values(by=['age'], inplace=True)
display(age_df.iloc[2])
```

### Question 5)
Please read the "shot1.csv" and "shot2.csv" files as a DataFrame object. Combine information of these two dataset into a single dataframe without any redundant column and display 10 records of the new dataframe.
Be careful combining these two datasets the new dataframe has redundant columns, find these columns and include only one copy of the redundant columns.

```python
# solution code
shot1_df = pd.read_csv("shot1.csv")
shot2_df = pd.read_csv("shot2.csv")

merged_df = pd.merge(shot1_df, shot2_df).drop(columns=['GAME_IDENTIFICATION'])
display(merged_df.head(10))
```

### Question 6)

Display first 10 records for the mean of shot numbers by player name by using obtained dataframe in Question 5.


```python
# solution code
display(merged_df.groupby(by=['player_name']).mean()['SHOT_NUMBER'].head(10))
```

### Question 7)
Write records stored in a DataFrame that is given to you as image to a SQLite database and read all data from SQL via a SQL query.

```python
conn = connect(':memory:') #Create the database in RAM
cursor = conn.cursor()
```

```python
# solution code
cursor.execute("CREATE TABLE table1(_index, gender, race, level_of_education, lunch, test_preparation, math_score, province)")
cursor.execute("""
        INSERT INTO table1 VALUES
            (0, 'female', 'group C', 'associate degree', 'free/reduced', 'none', 58, 'Washington'),
            (1, 'male', 'group A', 'high school', 'standart', 'none', 40, 'Georgia'),
            (2, 'female', 'group D', 'high school', 'free/reduced', 'none', 47, 'Mississippi'),
            (3, 'male', 'group E', 'college', 'free/reduced', 'completed', 70, 'Kentucky'),
            (4, 'male', 'group B', 'high school', 'standart', 'none', 38, 'Utah'),
            (5, 'male', 'group E', 'associate degree', 'standart', 'completed', 82, 'Washington'),
            (6, 'female', 'group D', 'high school', 'standart', 'none', 56, 'Oklahoma')
""")
conn.commit()
for row in cursor.execute("SELECT * FROM table1"):
    print(row)
```

### Question 8) 
Filter students whose level of education is "high school" and lunch group is "standard" using a SQL query

```python
# solution code
for row in cursor.execute("SELECT * FROM table1 WHERE level_of_education='high school' AND lunch='standart'"):
    print(row)
```

### Question 9)
Show province names in ascending order by using a SQL query

```python
# solution code
for row in cursor.execute("SELECT province FROM table1 ORDER BY province ASC"):
    print(row)
```

### Question 10)
Change the value of test preparation to "completed" for all records in which math score is higher than 55 and display it by using a SQL query

```python
# solution code
cursor.execute("UPDATE table1 SET test_preparation='completed' WHERE math_score > 55")
for row in cursor.execute("SELECT * FROM table1"):
    print(row)
```
